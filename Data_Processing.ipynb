### 3. Mengolah Dataset (30pts)

Dengan menggunakan dataset `application_processed.csv` buatlah dataframe baru dengan ketentuan:
1. Ambil semua users yang memiliki `email`. (ditandai dengan nilai pada column `flag_email` = 1) dan tidak memilik data `duplicates`. (5pts)
2. Buatlah column baru bernama `has_car_and_property` dimana nilainya merupakan range dari `0-1`. `1` apabila `flag_own_car` bernilai `1` dan `flag_own_property` bernilai `1`. `0` apabila terdapat atau semua nilai pada column `flag_own_car` dan `flag_own_property` terdapat angka 0. (5pts)
3. Ambil column yang diperlukan yaitu `id`, `gender`, `has_car_and_property`, `income_type`, `education_type`, `family_status`, `housing_type`, `occupation_type`, `amount_income_total`. (5pts)
4. Lakukan filter dengan ketentuan sebagai berikut: (5pts)
    - Memiliki `car` dan `property`
    - Tidak ada nilai null pada column `occupation_type`
    - Memiliki value pada column `occupation_type` dengan pola `staff`.
5. Sudah dilakukan sorting berdasarkan `income_type` dari `A-Z` dan `amount_income_total` dari `besar ke kecil`. (5pts)
6. Simpan kedalam bentuk `.csv` dengan nama `homework_3_result.csv`. (5pts)

**Expected Result:**

![Expected Result](assets/homework_exercise_3_result.png)

```python
## Your code below here
import pandas as pd

# Import data
application_processed = pd.read_csv('application_processed.csv')


#1 Semua users yang memiliki email dan tidak memilik data duplicates
email_users = application_processed[application_processed.flag_email == 1].drop_duplicates()

#2 Column baru
email_users['has_car_and_property'] = (email_users['flag_own_car'] 
                                       & email_users['flag_own_property']).astype(int)

#3 Mengambil column yang diperlukan
selected_columns = email_users[['id', 'gender', 'has_car_and_property', 'income_type', 'education_type', 'family_status', 'housing_type', 'occupation_type', 'amount_income_total']]
selected_columns

#4 Melakukan filter
filtered_columns = selected_columns[(selected_columns['has_car_and_property'] == 1)
                          & (selected_columns['occupation_type'].notnull()) 
                          & (selected_columns['occupation_type'].str.contains('staff'))]


#5 Melakukan sorting
sorting_columns = filtered_columns.sort_values(['income_type', 'amount_income_total'], ascending=[True, False])
sorting_columns.reset_index(drop=True)
```

```python
sorting_columns.to_csv("homework_3_result.csv", index=None)
```

### 5. Memanfaatkan Pivot Tabel (15pts)

Buatlah dataframe baru menggunakan fungsi pivot table untuk melihat `Embarked` apa saja yang `average_female_ticket_price` lebih besar daripada `average_male_ticket_price`. Informasi `average_female_ticket_price` dan `average_male_ticket_price` didapatkan dari rata-rata pada column `Fare`.

**Hint:**
1. menggunakan pivot table
2. dapat dilihat pada topik `DataFrame Aggregation` bagian `Pivot Table | Reshape Columns and Rows`

**Expected Result:**

![Expected Result](assets/homework_exercise_5_result.png)

```python
## Your code below here

# Import data
titanic_data = pd.read_csv('titanic.csv')

# Membuat pivot table
pd.pivot_table(titanic_data,
              index = ['Embarked', 'Sex'],
              values = ['Fare'],
              aggfunc = {'Fare' : ['mean']}
              )

# Merapihkan data / reshape
titanic_pivot = pd.pivot_table(titanic_data,
                          index = ['Embarked'],
                          columns = ['Sex'],
                          values = ['Fare']
                          )

titanic_pivot = titanic_pivot.reset_index()
titanic_pivot.columns = ['Embarked','average_female_ticket_price','average_male_ticket_price']
titanic_pivot

# Filtering 
titanic_pivot[titanic_pivot.average_female_ticket_price > titanic_pivot.average_male_ticket_price]
```

### 7. Merge DataFrame (25pts)

Buatlah 2 dataframe mengikuti petunjuk berikut:

df_1 :  Filter dataset 01 telecom_revenue.csv dengan multiple kondisi (5pts)
- MonthlyRevenue > 10 
- Occupation terdiri dari Professional, Student, and Crafts

df_2 :  Filter dataset 02 telecom_usage.csv dengan multiple kondisi (5pt)
- nilai UnansweredCalls > BlockedCalls

Jawablah pertanyaan berikut ini:
1. Hitunglah total CustomerID dan rata-rata DroppedCalls untuk masing-masing occupation. (10 Points)
2. Pada occupation apa, rata-rata DroppedCalls paling besar? (5 Points)

**Expected Result:**

![Expected Result](assets/homework_exercise_7_result.png)

```python
## Your code below here
import pandas as pd

# Import data
telecom_revenue_data = pd.read_csv('01 telecom_revenue.csv')



# Filtering
telecom_revenue_filtering = telecom_revenue_data[ (telecom_revenue_data.MonthlyRevenue>10)
                                                 &(telecom_revenue_data.Occupation.isin(['Professional','Student','Crafts']))]
telecom_revenue_filtering
```

```python
# Import data
telecom_usage_data = pd.read_csv('02 telecom_usage.csv')
telecom_usage_data

# Filtering
telecom_usage_filtering = telecom_usage_data[telecom_usage_data.UnansweredCalls > telecom_usage_data.BlockedCalls]
telecom_usage_filtering
```

```python
# Merge
merge_result = telecom_revenue_filtering.merge(telecom_usage_filtering,
                                          on = 'CustomerID',
                                         how = 'inner'
                                         )
merge_result
```

```python
# Rata-rata DroppedCalls paling besar
merge_final = merge_result.groupby('Occupation').agg({'CustomerID' : ['nunique'],
                                                      'DroppedCalls' : ['mean']
                                                     })
merge_final
```
